# kospeech 
- github:  https://github.com/sooftware/kospeech
- pytorch 기반의 딥러닝 모델로, 한국어만 지원
-  Deep Speech 2, Speech Transformer 등 다양한 알고리즘 기반의 모델을 지원
image.png

## End to End 모델
- 한국어의 STT 난이도가 높은 이유는, 수많은 자음 모음의 조합 뿐만 아니라 음소, 음절 등 언어에 고려해야하는 수많은 변수들이 있기 때문
- 2012년 이후 딥러닝이 발전하면서 해당 변수들에 대한 처리를 그냥 layer에 맡기는 것이 가능해졌고, 이에 음소나 음절의 처리, 문법, 발음 등을 모두 학습하게 하는 End to End 모델이 가능해 짐.

## 전처리를 지원
- github: https://github.com/sooftware/kospeech/tree/latest/dataset/kspon
- [AI Hub KsponSpeech](https://aihub.or.kr/aidata/105)로 알려진 1,000시간의 한국어 음성 코퍼스를 이용해 학습.
- KsponSpeech 말뭉치와 여러 모델(Deep Speech2, LAS, Transformer, Jasper, Conformer)에 대한 전처리 방법을 제안
- 음성인식 모델링이 어려운 대표적인 이유가 신호와 텍스트 간의 위치 정보를 맞추기 어렵다.

    - CTC(Connectionist temporal classification) 알고리즘과 Attention Algorithm 사용.
